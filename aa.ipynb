```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 创建一个图形
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制外层矩形框（系统功能）
ax.add_patch(patches.FancyBboxPatch((0.1, 0.8), 0.8, 0.15, boxstyle="round,pad=0.05", linewidth=2, edgecolor="black", facecolor="lightblue"))
ax.text(0.5, 0.875, '系统功能', horizontalalignment='center', verticalalignment='center', fontsize=12, fontweight='bold')

# 绘制子模块矩形框（用户管理功能、百科管理功能等）
modules = [
    ("用户管理功能", 0.1, 0.65),
    ("百科管理功能", 0.1, 0.5),
    ("用户社区管理功能", 0.1, 0.35),
    ("商城管理功能", 0.1, 0.2),
    ("用户反馈管理功能", 0.6, 0.65),
    ("数据库管理功能", 0.6, 0.5)
]

for module, x, y in modules:
    ax.add_patch(patches.FancyBboxPatch((x, y), 0.3, 0.1, boxstyle="round,pad=0.05", linewidth=2, edgecolor="black", facecolor="lightgreen"))
    ax.text(x + 0.15, y + 0.05, module, horizontalalignment='center', verticalalignment='center', fontsize=10, fontweight='bold')

# 绘制子功能矩形框
submodules = {
    "用户管理功能": [("用户注册", 0.1, 0.55), ("用户登录", 0.1, 0.45), ("密码管理", 0.1, 0.35)],
    "百科管理功能": [("词条管理", 0.1, 0.4), ("多语言支持", 0.1, 0.3), ("相关推荐", 0.1, 0.2)],
    "用户社区管理功能": [("帖子管理", 0.1, 0.25), ("帖子分类", 0.1, 0.15)],
    "商城管理功能": [("商家管理", 0.1, 0.1), ("商品管理", 0.1, 0.0)],
    "用户反馈管理功能": [("社区反馈", 0.6, 0.55), ("商城反馈", 0.6, 0.45)],
    "数据库管理功能": [("数据库连接", 0.6, 0.4), ("数据插入", 0.6, 0.3)]
}

for module, submodule_list in submodules.items():
    for submodule, x, y in submodule_list:
        ax.add_patch(patches.FancyBboxPatch((x, y), 0.3, 0.08, boxstyle="round,pad=0.05", linewidth=2, edgecolor="black", facecolor="lightyellow"))
        ax.text(x + 0.15, y + 0.04, submodule, horizontalalignment='center', verticalalignment='center', fontsize=9)

# 设置图形的显示范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')  # 关闭坐标轴

plt.title('系统功能层次方框图', fontsize=14)
plt.show()
```
